##중첩옵션(Compound Option)
중첩옵션(Compound Option)은 옵션을 기초 자산으로 하는 옵션을 의미합니다. 여기서는 간단한 중첩옵션의 가격을 계산하는 예제를 보여드리겠습니다. 중첩옵션의 두 가지 종류는 콜옵션을 기초 자산으로 하는 콜옵션(Call on Call)과 콜옵션을 기초 자산으로 하는 풋옵션(Put on Call)입니다.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# 블랙-숄즈 모델을 활용한 콜옵션 가격 계산 함수
def call_option_price(S, K, T, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    call_price = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    return call_price

# 중첩옵션의 가격 계산 함수 (Call on Call)
def compound_option_price(S, K1, K2, T1, T2, r, sigma, option_type='call'):
    """
    S: 기초 자산의 현재 가격
    K1: 첫 번째 옵션의 행사가격
    K2: 두 번째 옵션의 행사가격
    T1: 첫 번째 옵션의 만기 시간
    T2: 두 번째 옵션의 만기 시간
    r: 무위험 이자율
    sigma: 기초 자산의 변동성
    option_type: 'call' or 'put' (기초 자산이 되는 옵션의 종류)
    """
    # 첫 번째 옵션 가격 계산 (옵션이 되는 옵션)
    if option_type == 'call':
        option_price = call_option_price(S, K1, T1, r, sigma)
    elif option_type == 'put':
        option_price = put_option_price(S, K1, T1, r, sigma)

    # 중첩옵션 가격 계산
    compound_price = call_option_price(option_price, K2, T2, r, sigma)

    return compound_price

# 예시 데이터
S = 100  # 기초 자산의 현재 가격
K1 = 105 # 첫 번째 옵션의 행사가격
K2 = 10  # 두 번째 옵션의 행사가격
T1 = 1   # 첫 번째 옵션의 만기 기간 (1년)
T2 = 0.5 # 두 번째 옵션의 만기 기간 (0.5년)
r = 0.05 # 무위험 이자율 (5%)
sigma = 0.2  # 기초 자산의 변동성 (20%)

# 중첩옵션 가격 계산
compound_call_price = compound_option_price(S, K1, K2, T1, T2, r, sigma, option_type='call')
print(f"중첩옵션 (Call on Call) 가격: {compound_call_price:.2f}")

# 기초 자산 가격 변화에 따른 중첩옵션 가격 변화 시각화
S_values = np.linspace(50, 150, 100)
compound_call_prices = [compound_option_price(S, K1, K2, T1, T2, r, sigma, option_type='call') for S in S_values]

plt.figure(figsize=(10, 6))
plt.plot(S_values, compound_call_prices, label='중첩옵션 (Call on Call) 가격')
plt.xlabel('기초 자산의 가격 ($S$)')
plt.ylabel('옵션 가격')
plt.title('기초 자산 가격에 따른 중첩옵션 가격 변화')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 블랙-숄즈 모델을 활용한 콜옵션 가격 계산 함수
def call_option_price(S, K, T, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    call_price = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    return call_price

# 중첩옵션의 가격 계산 함수 (Call on Call)
def compound_option_price(S, K1, K2, T1, T2, r, sigma, option_type='call'):
    # 첫 번째 옵션 가격 계산 (옵션이 되는 옵션)
    if option_type == 'call':
        option_price = call_option_price(S, K1, T1, r, sigma)

    # 중첩옵션 가격 계산
    compound_price = call_option_price(option_price, K2, T2, r, sigma)

    return compound_price

# 실제 데이터 가져오기
# Apple Inc. 주가 데이터
apple = yf.Ticker("AAPL")
apple_data = apple.history(period="1y")
S_values = apple_data['Close'].values

# 기본 변수 설정
K1 = 150  # 첫 번째 옵션의 행사가격
K2 = 10   # 두 번째 옵션의 행사가격
T1 = 1    # 첫 번째 옵션의 만기 기간 (1년)
T2 = 0.5  # 두 번째 옵션의 만기 기간 (0.5년)
r = 0.05  # 무위험 이자율 (5%)
sigma = np.std(np.log(apple_data['Close'] / apple_data['Close'].shift(1)).dropna()) * np.sqrt(252)  # 기초 자산의 변동성

# 중첩옵션 가격 계산
compound_call_prices = [compound_option_price(S, K1, K2, T1, T2, r, sigma, option_type='call') for S in S_values]

# 시각화
plt.figure(figsize=(10, 6))
plt.plot(apple_data.index, compound_call_prices, label='중첩옵션 (Call on Call) 가격')
plt.xlabel('날짜')
plt.ylabel('옵션 가격')
plt.title('Apple 주가에 따른 중첩옵션 가격 변화')
plt.legend()
plt.grid(True)
plt.show()
```
