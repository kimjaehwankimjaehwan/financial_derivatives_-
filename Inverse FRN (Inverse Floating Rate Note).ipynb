##Inverse FRN (Inverse Floating Rate Note)
**Inverse FRN(Inverse Floating Rate Note)**은 변동 금리부 채권의 일종으로, 일반적인 변동 금리부 채권(FRN)과는 달리 시장 금리가 상승하면 지급 이자가 감소하고, 시장 금리가 하락하면 지급 이자가 증가하는 구조를 가진 채권입니다. 이는 투자자에게 금리 하락에 따른 혜택을 제공하며, 반대로 금리 상승 시에는 손실을 입게 됩니다.

##Inverse FRN의 특징

1. 금리 연동 구조:

  - Inverse FRN의 이자율은 일반적으로 기준 금리(예: LIBOR, SOFR)에서 특정 마진을 뺀 값으로 결정됩니다. 예를 들어, 이자율 = 고정 이자율 - (변동 기준 금리)이 됩니다.
2. 투자 목적:

  - Inverse FRN은 금리 하락 시 이자 수익을 증가시키기 위한 투자 도구로 사용됩니다. 금리가 하락하면 이자 수익이 증가하고, 금리가 상승하면 이자 수익이 감소합니다.
3. 위험 관리:

  - Inverse FRN은 금리 위험을 관리하거나, 금리 하락에 대비한 헤지 수단으로 사용될 수 있습니다.

```python
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt

# Google Colab에 NanumGothic 폰트 설치
!apt-get update -qq
!apt-get install fonts-nanum* -qq

# 설치된 폰트 확인 및 matplotlib에 적용
font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
font = fm.FontProperties(fname=font_path)
fm.fontManager.addfont(font_path)
plt.rc('font', family='NanumGothic')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Inverse FRN 파라미터
notional = 1000000  # 명목 금액
fixed_rate = 0.05  # 고정 이자율 (5%)
margin = 0.02  # 마진 (2%)
maturity = 5  # 만기 (5년)
payment_dates = np.arange(1, maturity + 1)  # 지불 날짜 (매년)

# 변동 기준 금리 (예: LIBOR)
np.random.seed(42)
base_rates = np.random.uniform(0.01, 0.05, maturity)  # 임의의 변동 금리 생성

# Inverse FRN의 이자율 계산
inverse_frn_rates = fixed_rate - base_rates - margin

# Inverse FRN의 이자 지급액 계산
interest_payments = notional * inverse_frn_rates

# 총 이자 지급액 계산
total_interest_payments = np.sum(interest_payments)

print(f"변동 기준 금리 (매년): {base_rates}")
print(f"Inverse FRN 이자율 (매년): {inverse_frn_rates}")
print(f"이자 지급액 (매년): {interest_payments}")
print(f"총 이자 지급액: {total_interest_payments:.2f} USD")

# 시각화
plt.figure(figsize=(12, 6))
plt.plot(payment_dates, base_rates, label='변동 기준 금리 (LIBOR)', marker='o')
plt.plot(payment_dates, inverse_frn_rates, label='Inverse FRN 이자율', marker='o')
plt.xlabel('연도')
plt.ylabel('이자율')
plt.title('Inverse FRN의 변동 기준 금리와 이자율')
plt.legend()
plt.grid(True)
plt.show()
```
